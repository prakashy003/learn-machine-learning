Let’s now take a look a scaling function offered by OpenCV
We can use this function to perform tasks such as zooming and shrinking images, which becomes vital in many image processing applications

```python
import cv2
import numpy as np
img = cv2.imread('cat.jpg')
res = cv2.resize(img,None,fx=0.1, fy=0.1, interpolation = cv2.INTER_AREA)
res1 = cv2.resize(img,None,fx=10, fy=10, interpolation = cv2.INTER_AREA)
cv2.imshow('shrinked image',res)
cv2.imshow('zoomed image',res1)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
